Analysing how each stock's price has changed over the past 5 years
- day
- week
- month
- year

```python
def extract_close(stock, interval):
    date_lst = []
    close_lst = []
    close_df = stock.history(period="5y")["Close"]
    for i in range(close_df.shape[0]):
        dt = close_df.index[i].date() 
        if interval == 'day':
            date_lst.append(dt)
            close_lst.append(close_df[i])
        elif interval == 'week' and i % 5 == 0:
            date_lst.append(dt)
            close_lst.append(close_df[i])
        elif interval == 'month' and dt.day == 15: # use the most recent day
            date_lst.append(dt)
            close_lst.append(close_df[i])
    return date_lst, close_lst
```

```python
import matplotlib.pyplot as plt
def historical_price_plot(company):
    stock = yf.Ticker(company)
    stock_date_d, stock_close_d = extract_close(stock, interval="day")
    stock_date_w, stock_close_w = extract_close(stock, interval="week")
    stock_date_m, stock_close_m = extract_close(stock, interval="month")

    fig, axs = plt.subplots(3)

    axs[0].set_title(f"{company} Closing Price")
    axs[0].plot(stock_date_d, stock_close_d)
    axs[0].set_xlabel("Date")
    axs[0].set_ylabel("Closing Price")

    axs[1].plot(stock_date_w, stock_close_w)
    axs[1].set_xlabel("Date")
    axs[1].set_ylabel("Closing Price")

    axs[2].plot(stock_date_m, stock_close_m)
    axs[2].set_xlabel("Date")
    axs[2].set_ylabel("Closing Price")

    plt.show()
```

```python
historical_price_plot("AAPL")
```
