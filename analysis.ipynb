Analysing how each stock's price has changed over the past 5 years
- day
- week
- month
- year

```python
def extract_close(stock, interval):
    date_lst = []
    close_lst = []
    close_df = stock.history(period="5y")["Close"]
    for i in range(close_df.shape[0]):
        dt = close_df.index[i].date() 
        if interval == 'day':
            date_lst.append(dt)
            close_lst.append(close_df[i])
        elif interval == 'week' and i % 5 == 0:
            date_lst.append(dt)
            close_lst.append(close_df[i])
        elif interval == 'month' and dt.day == 15: # use the most recent day
            date_lst.append(dt)
            close_lst.append(close_df[i])
    return date_lst, close_lst
```

```python
import matplotlib.pyplot as plt
def historical_price_plot(company):
    stock = yf.Ticker(company)
    stock_date_d, stock_close_d = extract_close(stock, interval="day")
    stock_date_w, stock_close_w = extract_close(stock, interval="week")
    stock_date_m, stock_close_m = extract_close(stock, interval="month")

    fig, axs = plt.subplots(3)

    axs[0].set_title(f"{company} Closing Price")
    axs[0].plot(stock_date_d, stock_close_d)
    axs[0].set_xlabel("Date")
    axs[0].set_ylabel("Closing Price")

    axs[1].plot(stock_date_w, stock_close_w)
    axs[1].set_xlabel("Date")
    axs[1].set_ylabel("Closing Price")

    axs[2].plot(stock_date_m, stock_close_m)
    axs[2].set_xlabel("Date")
    axs[2].set_ylabel("Closing Price")

    plt.show()
```

The historical price of some stocks seem to have the same trend. Plot the historical prices of different stocks on the same axes (weekly).

```python
def create_lag_close_volume(df, days):
    df = df.copy()
    df["Lag_Close"+str(days)] = df["Close"].shift(days)
    df["Close_Change"+str(days)] = df["Close"] - df["Lag_Close"+str(days)]
    df["Lag_Volume"+str(days)] = df["Volume"].shift(days)
    df["Volume_Change"+str(days)] = df["Volume"] - df["Lag_Volume"+str(days)]

    return df
```

I remember reading about how an increase in volume could mark the start of a breakthrough. I would like to see the correlation between the volume and change in price.

```python
from sklearn.linear_model import LinearRegression
def volume_price_change_visualisation(stock, days):
    ticker = yf.Ticker(stock)
    lag_cv = create_lag_close_volume(ticker.history(period="5y"), days).iloc[1:]
    
    # visualisations
    fig, ax = plt.subplots(2, 1, figsize=(10,10), sharex="col")


    x = lag_cv[[f"Lag_Volume{days}"]].values.reshape(-1,1)
    y = lag_cv[f"Close_Change{days}"]
    y_abs = abs(y)

    def plot_scatter_regression(x, y, i):
        # scatter plot and regression line
        ax[i].scatter(x, y)
        # linear regression
        reg = LinearRegression()
        reg.fit(x, y)
        y_pred = reg.predict(x)
        ax[i].plot(x, y_pred, c="red")

        if i == 0:
            ax[i].set_ylabel("Price Change")
        else:
            ax[i].set_ylabel("Absolute Price Change")

    plot_scatter_regression(x, y, 0)
    plot_scatter_regression(x, y_abs, 1)
    ax[1].set_xlabel("Volume Change")
    ax[0].set_title("Relationship between Volume Change and Price Change")

    plt.show()
```

```python
volume_price_change_visualisation("AAPL", 1)
```
